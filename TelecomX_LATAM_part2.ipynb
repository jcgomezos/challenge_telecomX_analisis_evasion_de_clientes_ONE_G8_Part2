```python
import pandas as pd

data = pd.read_csv('/content/datos_tratados.csv')
data.head()
```

```python
data = data.drop(columns=['customerID'], errors='ignore')
data.head()
```

```python
columnas = data.columns
```

```python
columnas
```

```python
data.dtypes
```

```python
categoricas = ['Churn', 'customer.gender','customer.Partner', 'customer.Dependents',
               'phone.PhoneService', 'phone.MultipleLines', 'internet.InternetService',
               'internet.OnlineSecurity', 'internet.OnlineBackup',
               'internet.DeviceProtection', 'internet.TechSupport',
               'internet.StreamingTV', 'internet.StreamingMovies', 'account.Contract',
               'account.PaperlessBilling', 'account.PaymentMethod']
data_codificada= pd.get_dummies(data=data, columns=categoricas, dtype=int)
data_codificada.sample(5)
```

```python
# Conteo de clientes por clase (0 = No canceló, 1 = Canceló)
conteo = data_codificada['Churn_Yes'].value_counts()

# Proporción relativa
proporcion = data_codificada['Churn_Yes'].value_counts(normalize=True)

print("📊 Conteo de clientes:")
print(conteo)

print("\n📈 Proporción de clientes:")
print(proporcion)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter


# Separar variables predictoras (X) y variable objetivo (y)
X = data_codificada.drop("Churn_Yes", axis=1)
y = data_codificada["Churn_Yes"]

# Ver distribución original
print("Distribución original:", Counter(y))

# Crear el objeto SMOTE
smote = SMOTE(random_state=42)

# Aplicar SMOTE
X_resampled, y_resampled = smote.fit_resample(X, y)

# Ver distribución balanceada
print("Distribución balanceada:", Counter(y_resampled))

# Convertir a DataFrame nuevamente
data_balanceada = pd.DataFrame(X_resampled, columns=X.columns)
data_balanceada["Churn_Yes"] = y_resampled

# Revisar primeras filas
data_balanceada.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular la matriz de correlación
corr_matrix = data_balanceada.corr()

# Filtrar solo la correlación con la variable objetivo 'Churn_Yes'
correlaciones_churn = corr_matrix["Churn_Yes"].sort_values(ascending=False)

print("🔍 Correlaciones con Churn_Yes:")
print(correlaciones_churn)

# Graficar mapa de calor completo
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Matriz de correlación - Todas las variables", fontsize=16)
plt.show()

# Graficar solo las 10 variables más correlacionadas con Churn_Yes
top_corr = correlaciones_churn[1:11]  # Excluye la propia Churn_Yes
plt.figure(figsize=(8, 6))
sns.barplot(x=top_corr.values, y=top_corr.index, palette="coolwarm")
plt.title("Top 10 variables más correlacionadas con Churn_Yes", fontsize=14)
plt.xlabel("Correlación")
plt.ylabel("Variable")
plt.show()
```

```python

# Ajuste de estilo
sns.set(style="whitegrid")

# Tiempo de contrato vs Cancelación (boxplot)
plt.figure(figsize=(7, 5))
sns.boxplot(x="Churn_Yes", y="customer.tenure", data=data_balanceada, palette="Set2")
plt.title("Tiempo de contrato vs Cancelación", fontsize=14)
plt.xlabel("Cancelación (1 = Sí, 0 = No)")
plt.ylabel("Tiempo de contrato (meses)")
plt.show()

# Gasto total vs Cancelación (boxplot)
plt.figure(figsize=(7, 5))
sns.boxplot(x="Churn_Yes", y="account.Charges.Total", data=data_balanceada, palette="Set2")
plt.title("Gasto total vs Cancelación", fontsize=14)
plt.xlabel("Cancelación (1 = Sí, 0 = No)")
plt.ylabel("Gasto total")
plt.show()

# Tiempo de contrato vs Gasto total (scatter plot, coloreado por churn)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="customer.tenure", y="account.Charges.Total", hue="Churn_Yes", data=data_balanceada, palette="Set1", alpha=0.6)
plt.title("Tiempo de contrato vs Gasto total (color = Cancelación)", fontsize=14)
plt.xlabel("Tiempo de contrato (meses)")
plt.ylabel("Gasto total")
plt.legend(title="Churn")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Variables independientes (X) y variable objetivo (y)
X = data_balanceada.drop(["Churn_Yes", "Churn_No"], axis=1)  # Todas menos la de churn
y = data_balanceada["Churn_Yes"]               # Variable objetivo

# División 70% entrenamiento, 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Verificación de tamaños
print(f"Tamaño entrenamiento: {X_train.shape[0]} registros")
print(f"Tamaño prueba: {X_test.shape[0]} registros")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Pipeline con normalización y modelo
logreg_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(max_iter=1000, random_state=42))
])

# Entrenar
logreg_pipeline.fit(X_train, y_train)

# Predicciones
y_pred_logreg = logreg_pipeline.predict(X_test)

# Métricas
print("=== Regresión Logística ===")
print(confusion_matrix(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))
print("Accuracy:", accuracy_score(y_test, y_pred_logreg))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Entrenar modelo
rf_model = RandomForestClassifier(
    n_estimators=100,     # número de árboles
    max_depth=5,          # profundidad máxima de cada árbol
    random_state=42
)
rf_model.fit(X_train, y_train)

# Predicciones
y_pred_rf = rf_model.predict(X_test)

# Métricas
print("\n=== Random Forest ===")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

#**Conclusión**


A lo largo del desarrollo, construimos y evaluamos dos modelos de clasificación para predecir la pérdida de clientes (churn): Regresión Logística y Random Forest. El objetivo fue comparar su desempeño sin realizar ajuste de hiperparámetros, partiendo de una preparación correcta de los datos para evitar fugas de información (por ejemplo, eliminando la variable Churn_No que contenía información redundante sobre la variable objetivo).

En la evaluación, la Regresión Logística presentó un desempeño más balanceado, con una exactitud del 85.2%, precisión y recall similares para ambas clases, lo que indica una buena capacidad para identificar tanto a clientes que se quedan como a los que se van. Esto la convierte en un modelo sencillo, interpretable y confiable para un primer acercamiento.

El Random Forest, con una exactitud del 81.3%, mostró un mejor recall para la clase positiva (clientes que se van), pero a costa de una menor precisión para la clase negativa. Esto sugiere que tiende a identificar más posibles casos de churn, pero también genera más falsos positivos. Aunque es más flexible y capaz de capturar relaciones no lineales, en este caso no superó a la Regresión Logística posiblemente por falta de ajuste de hiperparámetros y por una configuración inicial que podría estar favoreciendo el sobreajuste parcial a ciertos patrones del entrenamiento.

En conjunto, el ejercicio permitió:

Identificar y corregir fugas de información.

Comprender cómo se comportan dos enfoques diferentes: uno lineal e interpretable, y otro basado en ensambles y no linealidad.

Analizar métricas más allá de la exactitud, evaluando la precisión, recall, F1-score y matrices de confusión para obtener una visión más completa del rendimiento.

En esta etapa, la Regresión Logística se muestra como el mejor punto de partida, mientras que el Random Forest tiene potencial para mejorar con un ajuste adecuado de hiperparámetros (profundidad, número de árboles, tamaño mínimo de hojas, etc.) y una validación cruzada más exhaustiva.
